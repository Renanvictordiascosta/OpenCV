```python
# Importação das bibliotecas
import numpy as np
import cv2
import matplotlib.pyplot as plot
```

```python
# Mudar configuração de cores e mostrar a imagem, com um título
def showImage(img):
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plot.imshow(img)
  plot.title("METAGROSS BOLADO, BY PENTA")
  plot.show()
```

```python
# Demonstrar escala de cor em pontos da imagem
def Count(height, width, img):
  for y in range(0, height):
    for x in range(0, width):
      if x == y:
        print(f"Posição ({x}, {y}) possui escala de cor {img[y][x]}")
```

```python
# Mostrar a escala da imagem e quantos canais de cores há nela
def Dimentions(img):
  height, width, color_channels = img.shape
  print(f"Dimensão: {width} x {height}")
  print(f"Canais de Cores: {color_channels}")
  Count(height, width, img)
```

```python
# Reunir as funções anteriores em uma principal "main"
def main():
    img = cv2.imread("bestpokemon.webp")
    Dimentions(img)
    showImage(img)
```

```python
# Chamada da "main"
main()
```
